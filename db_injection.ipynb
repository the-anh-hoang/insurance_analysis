```python
total_expenditures_in_real_dollars = pd.read_csv('nhe23-tables/total_expenditure_in_real_dolars_2017.csv')
print(total_expenditures_in_real_dollars.columns)
total_expenditures_in_real_dollars.head()
```

```python
df = total_expenditures_in_real_dollars
df.replace(' - ', np.nan, inplace=True)  # Replace '-' with NaN
df['National_Health_Expenditures_Billions'] = (
    df['National_Health_Expenditures_Billions']
    .str.replace('$', '')  # Remove dollar signs
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to float
)
df['Personal_Health_Care_Billions'] = (
    df['Personal_Health_Care_Billions']
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to float
)

# Result
df
```
